```python
from torch.utils.data import DataLoader, Dataset
import torch
from zipfile import ZipFile
import os
import numpy as np
import json
import matplotlib.pyplot as plt
import cv2
import io
from PIL import Image
```

```python
def show_box2d(img, labels):
    h, w, _ = img.shape
    for label in labels:
        cx, cy = int(label[1]*w), int(label[2]*h)
        lx, ly = int(label[3]*w), int(label[4]*h)
        start_point = (cx-int(lx/2), cy-int(ly/2))
        end_point = (cx+int(lx/2), cy+int(ly/2))
        color = (255, 0, 0)
        thickness = 10
        print(start_point, end_point)
        img = cv2.rectangle(img, start_point, end_point, color, thickness)
    plt.imshow(img)
    plt.pause(0.001)
```

```python
label_path = "./labels/train.txt"
labels = np.loadtxt(label_path).reshape(-1, 5)
labels
```

```python
image_path = "./images/train.jpg"
img = cv2.imread(image_path)
show_box2d(img, labels)
```
